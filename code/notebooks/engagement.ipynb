```python
cols = [
    'CARBON_EMISSIONS_TARGET_BASE_YEAR',
    'CARBON_EMISSIONS_TARGET_REDUCT_PCT',
    'CARBON_EMISSIONS_TARGET_YEAR'
]

plt.figure(figsize=(16, 4))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)

    # Appliquer les filtres selon la colonne
    if col == 'CARBON_EMISSIONS_TARGET_BASE_YEAR':
        data_filtered = data[data[col] <= 2025]
    elif col == 'CARBON_EMISSIONS_TARGET_YEAR':
        data_filtered = data[data[col] <= 2060]
    else:
        data_filtered = data

    sns.histplot(data_filtered[col].dropna(), kde=True, bins=20)
    plt.title(col)
    plt.xlabel("")
    plt.ylabel("FrÃ©quence")

plt.tight_layout()
plt.show()
```

```python
# Filtrage des donnÃ©es valides
df_plot = data[
    (data['CARBON_EMISSIONS_TARGET_BASE_YEAR'] <= 2025) &
    (data['CARBON_EMISSIONS_TARGET_YEAR'] <= 2060)
][['CARBON_EMISSIONS_TARGET_BASE_YEAR', 'CARBON_EMISSIONS_TARGET_YEAR']].dropna()

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_plot,
    x='CARBON_EMISSIONS_TARGET_BASE_YEAR',
    y='CARBON_EMISSIONS_TARGET_YEAR',
    alpha=0.6
)
plt.title("Objectifs carbone : annÃ©e de base vs annÃ©e cible\n(filtrÃ© jusqu'Ã  2025 et 2060)")
plt.xlabel("AnnÃ©e de base")
plt.ylabel("AnnÃ©e cible")
plt.grid(True)
plt.show()
```

```python
frequency_table = pd.DataFrame({
    'HAS_COMMITTED_TO_SBTI_TARGET': [
        data['HAS_COMMITTED_TO_SBTI_TARGET'].isna().sum(),       
        (data['HAS_COMMITTED_TO_SBTI_TARGET'] == True).sum(),    
        (data['HAS_COMMITTED_TO_SBTI_TARGET'] == False).sum()  
    ],
    'HAS_SBTI_APPROVED_TARGET': [
        data['HAS_SBTI_APPROVED_TARGET'].isna().sum(),            
        (data['HAS_SBTI_APPROVED_TARGET'] == True).sum(),       
        (data['HAS_SBTI_APPROVED_TARGET'] == False).sum()        
    ]
}, index=['NaN', 'True', 'False'])

print(frequency_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir 'HAS_SBTI_APPROVED_TARGET' en chaÃ®ne de caractÃ¨res ('nan', 'true', 'false')
data['HAS_SBTI_APPROVED_TARGET'] = data['HAS_SBTI_APPROVED_TARGET'].apply(
    lambda x: 'nan' if pd.isna(x) else ('true' if x else 'false')
)

# Filtrer les donnÃ©es en fonction de la colonne 'HAS_COMMITTED_TO_SBTI_TARGET'
data_committed_nan = data[data['HAS_COMMITTED_TO_SBTI_TARGET'].isna()]
data_committed_false = data[data['HAS_COMMITTED_TO_SBTI_TARGET'] == False]
data_committed_true = data[data['HAS_COMMITTED_TO_SBTI_TARGET'] == True]

# CrÃ©er les histogrammes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Histogramme pour les valeurs oÃ¹ 'HAS_COMMITTED_TO_SBTI_TARGET' est NaN
axs[0].hist(data_committed_nan['HAS_SBTI_APPROVED_TARGET'], bins=3, color='blue', alpha=0.7, width=0.6)
axs[0].set_title('HAS_COMMITTED_TO_SBTI_TARGET = NaN')
axs[0].set_xlabel('HAS_SBTI_APPROVED_TARGET')
axs[0].set_ylabel('Frequency')

# Histogramme pour les valeurs oÃ¹ 'HAS_COMMITTED_TO_SBTI_TARGET' est False
axs[1].hist(data_committed_false['HAS_SBTI_APPROVED_TARGET'], bins=3, color='green', alpha=0.7, width=0.6)
axs[1].set_title('HAS_COMMITTED_TO_SBTI_TARGET = False')
axs[1].set_xlabel('HAS_SBTI_APPROVED_TARGET')
axs[1].set_ylabel('Frequency')

# Histogramme pour les valeurs oÃ¹ 'HAS_COMMITTED_TO_SBTI_TARGET' est True
axs[2].hist(data_committed_true['HAS_SBTI_APPROVED_TARGET'], bins=3, color='red', alpha=0.7, width=0.6)
axs[2].set_title('HAS_COMMITTED_TO_SBTI_TARGET = True')
axs[2].set_xlabel('HAS_SBTI_APPROVED_TARGET')
axs[2].set_ylabel('Frequency')

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Optionnel: ajuster manuellement les marges pour plus d'espace
# plt.subplots_adjust(wspace=0.4)

# Afficher les histogrammes
plt.show()
```
